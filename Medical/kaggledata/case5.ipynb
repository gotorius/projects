パワポ用の医療画像を何枚か出力するプログラム

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# 画像ディレクトリのパス
img_dir = '/mnt/data1/gotou/Medical/b4us/kaggledata/train'
img_files = [f for f in os.listdir(img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.tif'))]

# 最初の5枚を表示
n_show = 5
plt.figure(figsize=(15, 3))
for i, fname in enumerate(img_files[:n_show]):
    img_path = os.path.join(img_dir, fname)
    img = Image.open(img_path)
    plt.subplot(1, n_show, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(fname)
    plt.axis('off')
plt.suptitle('データセットの画像サンプル')
plt.show()
```
